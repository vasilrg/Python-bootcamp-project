```python
# Import of libraries needed for the visualisations

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# Load of the required csv file

df = pd.read_csv("insurance.csv")
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
# Scattterplot of the insurance data 

plt.scatter(data=df, x = 'age', y = 'charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
X = df[['age']].values
y = df[['charges']].values
```

```python
# Call of the LinearRegression function and assigining it to prediction model age_affects_ins

age_affects_ins = LinearRegression()
age_affects_ins.fit(X,y)
```

```python
# Declaration of a prediction value for the charges and visualising Best-fit line 

y_pred = age_affects_ins.predict(X)

plt.scatter(X, y , color = 'b')
plt.plot(X, y_pred, color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```
